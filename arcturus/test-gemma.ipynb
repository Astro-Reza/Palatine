### LIVE ANSWER CODE

To get the model to print its output live (like a typing effect) instead of waiting for the entire response to finish, you need to enable streaming.

If you want to see the model's "thoughts" (its internal reasoning process) before it gives the final answer, you will also need to point it to a specific "Thinking" model.

```python
import os
from dotenv import load_dotenv
from google import genai
import sys

# 1. Setup
api_key = os.getenv("GEMMA_API_KEY")
if not api_key:
    raise ValueError("GEMMA_API_KEY not found in environment variables. Check your .env file.")

client = genai.Client(api_key=api_key)
model_id = 'gemma-4-26b-a4b-it'

print(f"Using model: {model_id}")
print("Sending request to Google's servers...\n")
print("--- Live Output Begins ---")

# 2. Generate content with streaming
try:
    response = client.models.generate_content_stream(
        model=model_id,
        contents="How does a PID controller adjust motor speed? Walk me through the math.", 
    )

    # 3. Iterate through the chunks as they arrive from the server
    for chunk in response:
        if chunk.text:
            print(chunk.text, end="", flush=True)
except Exception as e:
    print(f"\nError during API request: {e}")

print("\n\n--- Output Complete ---")
```

```python
import os
from dotenv import load_dotenv
from google import genai
from google.genai import types

load_dotenv()
api_key = os.getenv("GEMMA_API_KEY")
if not api_key:
    raise ValueError("GEMMA_API_KEY not found in environment variables.")

client = genai.Client(api_key=api_key)
model_id = "gemma-4-26b-a4b-it"
print(f"Using model: {model_id}")
print("Sending request...\n")

thoughts = ""
answer = ""

try:
    for chunk in client.models.generate_content_stream(
        model=model_id,
        contents="How does a PID controller adjust motor speed? Walk me through the math.",
        config=types.GenerateContentConfig(
            thinking_config=types.ThinkingConfig(thinking_level="high")  # FIX 1: use ThinkingConfig
        ),
    ):
        for part in chunk.candidates[0].content.parts:  # FIX 2: iterate parts, check part.thought
            if not part.text:
                continue
            if part.thought:
                if not thoughts:
                    print("=" * 40)
                    print("[THINKING PROCESS]")
                    print("=" * 40)
                print(part.text, end="", flush=True)
                thoughts += part.text
            else:
                if not answer:
                    print("\n\n" + "=" * 40)
                    print("[ACTUAL RESPONSE]")
                    print("=" * 40)
                print(part.text, end="", flush=True)
                answer += part.text

except Exception as e:
    print(f"\nError: {e}")

print("\n\n--- Output Complete ---")
```

### Dynamic Model Routing (Fast -> Heavy)

This cell uses `gemini-3.1-flash-lite` as a low-cost router and forwards complex prompts to `gemma-4-26b-a4b-it`.
The two destination model IDs are defined once at the top so the flow stays simple and maintainable.

```python
import os
from pydantic import BaseModel, Field
from dotenv import load_dotenv
from google import genai
from google.genai import types

# Load environment once
load_dotenv(dotenv_path="../.env")
api_key = os.getenv("GEMMA_API_KEY")
if not api_key:
    raise ValueError("GEMMA_API_KEY not found in environment variables.")

client = genai.Client(api_key=api_key)

# Define models once at the top
MODEL_FAST = "models/gemini-3.1-flash-lite"
MODEL_HEAVY = "models/gemma-4-26b-a4b-it"

class RouterDecision(BaseModel):
    complexity: str = Field(
        description="LOW for simple prompts, HIGH for multi-step math/code/engineering."
    )
    selected_model: str = Field(
        description=f"Must be either {MODEL_FAST} or {MODEL_HEAVY}."
    )
    reasoning: str = Field(
        description="One short sentence on why this route was selected."
    )

def route_prompt(user_prompt: str) -> RouterDecision:
    router_prompt = f"""Analyze the user prompt and choose routing output.
- Choose '{MODEL_FAST}' for greetings, simple Q&A, rewrite, summary, or short explanation.
- Choose '{MODEL_HEAVY}' for multi-step math, coding, engineering analysis, or deep reasoning.

Return strict JSON matching the schema.

User Prompt: {user_prompt}
"""

    try:
        route_response = client.models.generate_content(
            model=MODEL_FAST,
            contents=router_prompt,
            config=types.GenerateContentConfig(
                response_mime_type="application/json",
                response_schema=RouterDecision,
                temperature=0.0,
            ),
        )
        decision = RouterDecision.model_validate_json(route_response.text)
    except Exception as e:
        print(f"Routing failed ({e}). Falling back to fast model.")
        decision = RouterDecision(
            complexity="LOW",
            selected_model=MODEL_FAST,
            reasoning="Fallback to fast model due to routing error.",
        )

    # Guardrail for unexpected model names
    if decision.selected_model not in {MODEL_FAST, MODEL_HEAVY}:
        decision.selected_model = MODEL_FAST
        decision.complexity = "LOW"
        decision.reasoning = "Router returned invalid model; forced fast fallback."

    return decision

def dynamic_route_and_execute(user_prompt: str):
    print(f"\n[Incoming Prompt] {user_prompt}")
    decision = route_prompt(user_prompt)
    print(
        f"[Router] complexity={decision.complexity} | model={decision.selected_model}\n"
        f"Reason: {decision.reasoning}"
    )

    config_kwargs = {}
    if decision.selected_model == MODEL_HEAVY:
        config_kwargs["thinking_config"] = types.ThinkingConfig(thinking_level="high")

    print(f"\n[Streaming via] {decision.selected_model}\n")
    thoughts_started = False
    answer_started = False

    try:
        stream = client.models.generate_content_stream(
            model=decision.selected_model,
            contents=user_prompt,
            config=types.GenerateContentConfig(**config_kwargs),
        )

        for chunk in stream:
            candidates = getattr(chunk, "candidates", None)
            if not candidates:
                continue
            content = getattr(candidates[0], "content", None)
            parts = getattr(content, "parts", None) if content else None
            if not parts:
                continue

            for part in parts:
                text = getattr(part, "text", None)
                if not text:
                    continue

                if getattr(part, "thought", False):
                    if not thoughts_started:
                        print("=" * 40)
                        print("[THINKING PROCESS]")
                        print("=" * 40)
                        thoughts_started = True
                    print(text, end="", flush=True)
                else:
                    if not answer_started:
                        print("\n\n" + "=" * 40)
                        print("[ACTUAL RESPONSE]")
                        print("=" * 40)
                        answer_started = True
                    print(text, end="", flush=True)

    except Exception as e:
        print(f"\nExecution error: {e}")
    finally:
        print("\n\n" + "-" * 30 + " End of Stream " + "-" * 30)

# Quick test prompts
dynamic_route_and_execute("Hey there! Can you introduce yourself briefly?")
dynamic_route_and_execute("How does a PID controller adjust motor speed? Walk me through the math.")
```
